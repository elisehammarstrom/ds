A courier company operates a fleet of delivery trucks that make deliveries to different parts of the city. The trucks are equipped with GPS tracking devices that record the location of each truck at regular intervals. The locations are divided into three regions: downtown, the suburbs, and the countryside, however there is always the possibility the truck breaks down and it goes to the workshop. The following table shows the probabilities of a truck transitioning between these regions at each time step:

| Current region | Probability of transitioning to downtown | Probability of transitioning to the suburbs | Probability of transitioning to the countryside | Probability of transitioning to the Workshop |
|-|-|-|-|-|
| Downtown       | 0.3 | 0.7 | 0 | 0 |
| Suburbs        | 0.2 | 0.5 | 0.3 | 0 |
| Countryside    | 0 | 0 | 0.5 | 0.5 |
| Workshop    | 0 | 0 | 0 | 1 |

1. If a truck is currently in the downtown, what is the probability that it will be in the countryside region after 10 time steps? [2p]
2. If a truck is currently in the downtown, what is the probability that it will be in the countryside region **the first time** after three time steps or more? [2p]
3. Is this Markov chain irreducible? Explain your answer. [3p]
4. What is the stationary distribution? Furthermore it it reversible? (Explain your answer) [3p]
5. Advanced question: What is the expected number of steps it takes starting from the Downtown region to first reach the Workshop? Hint: to get within 1 decimal point, it is enough to compute the probabilities for hitting times below 50. Motivate your answer in detail [4p]. You could also solve this question by simulation, but this gives you a maximum of [2p].



```python
# Define the transition matrix
transition_matrix = np.array([
    [0.3, 0.7, 0.0, 0.0],  # Downtown to other regions
    [0.2, 0.5, 0.3, 0.0],  # Suburbs to other regions
    [0.0, 0.0, 0.5, 0.5],  # Countryside to other regions
    [0.0, 0.0, 0.0, 1.0]   # Workshop (self-loop)
])
```

```python
# Part 1

# Fill in the answer to part 1 below
problem1_p1 = np.linalg.matrix_power(transition_matrix, 10)[0, 2]
print(problem1_p1)
```

```python
# Part 2

# Fill in the answer to part 2 below
cumulative_probability_countryside = 0
for step in range(3, 51):  # Up to 50 steps
    state_before_step = np.linalg.matrix_power(transition_matrix, step - 1)[0, :]
    probability_to_countryside_this_step = state_before_step[1] * transition_matrix[1, 2] + state_before_step[0] * transition_matrix[0, 2]
    cumulative_probability_countryside += probability_to_countryside_this_step
problem1_p2 = cumulative_probability_countryside
print(problem1_p2)
```

```python
# Part 6

# Put the code for part 6 below this line
def makeEDF(data):
    """
    Creates an empirical distribution function from the given data.
    """
    n = len(data)
    sorted_data = np.sort(data)
    cum_freq = np.arange(1, n+1) / n
    return np.column_stack((sorted_data, cum_freq))

edf = makeEDF(residuals)

# Using the provided plotEDF function to plot the EDF with confidence bands
plotEDF(edf, points_at_jump=True, confidence_band=True, alpha=0.95)
```

Identification of Outliers or Anomalies: If certain portions of the EDF fall outside the confidence band, this could indicate the presence of outliers or anomalies in the data that the model is not accounting for. ## Random variable generation

1. [4p] Using inversion sampling, construct 1000 samples from the below distribution
$$
    F[x] = 
    \begin{cases}
        0, & x \leq 0 \\
        e^x-1, & 0 < x < \ln(2) \\
        1, & x \geq \ln(2)
    \end{cases}
$$
2. [2p] Use the above 1000 samples to estimate the mean and variance
3. [4p] Using the **Accept-Reject** sampler (**Algorithm 1** in TFDS notes) construct 1000 samples from the same distribution, what proposal distribution did you choose and why? What proportion of samples where accepted?
4. [3p] Explain if it is possible to sample from the density
$$
    f(x) = C e^{-(x^2-2)^2}
$$
using the **Accept-Reject** sampler (**Algorithm 1** in TFDS notes) with sampling density given the Gaussian. Here $C$ is a constant to make sure that $f$ is a density, it between roughly $1.34$ and $1.35$.

```python
# Part 1

# Write your code below
import numpy as np

# Define the distribution function F(x)
def F(x):
    return np.where(x <= 0, 0, np.where(x < np.log(2), np.exp(x) - 1, 1))

# Inversion method: create a function to invert F(x)
def inverse_F(u):
    return np.log(1 + u)

# Generate 1000 uniform samples in the range [0, 1]
u = np.random.uniform(0, 1, 1000)

# Apply the inverse function to the uniform samples
samples = inverse_F(u)

samples[:10]  # Display first 10 samples for a quick check
```

```python
# Part 1

# Put the resulting samples into the following variable

problem3_samples = [0.36221768, 0.421716  , 0.23296831, 0.48114423, 0.62941564,
       0.54935226, 0.28163577, 0.27495194, 0.66206109, 0.32714268]
```

```python
# Part 2

# Calculate the mean and variance of the generated samples
problem3_mean = np.mean(problem3_samples)
problem3_variance = np.var(problem3_samples)
problem3_mean, problem3_variance
```

```python
# Part 3

# Write your code to solve the problem below

def target_density(x):
    """ Target density function derived from F[x]. """
    return np.where((x > 0) & (x < np.log(2)), np.exp(x), 0)

# Define the bounds of the uniform proposal distribution
a, b = 0, np.log(2)

# Maximum value of the target density within the interval [a, b]
max_target_density = max(target_density(np.linspace(a, b, 1000)))

# Initialize variables for Accept-Reject sampling
samples_accept_reject = []
total_samples = 0
accepted_samples = 0

# Generate 1000 samples using Accept-Reject sampler
while len(samples_accept_reject) < 1000:
    # Sample from the proposal uniform distribution
    x_proposal = np.random.uniform(a, b)
    
    # Sample from uniform distribution for acceptance
    u = np.random.uniform(0, max_target_density)
    
    # Increase total sample count
    total_samples += 1

    # Accept or reject the sample
    if u <= target_density(x_proposal):
        samples_accept_reject.append(x_proposal)
        accepted_samples += 1

# Proportion of accepted samples
proportion_accepted = accepted_samples / total_samples

proportion_accepted, samples_accept_reject[:10]  # Display the acceptance proportion and first 10 samples
```

```python
# Part 3

# Samples from the Accept-Reject method
problem3_samples_accept_reject = samples_accept_reject

# Proportion of samples accepted
problem3_acceptance_rate = proportion_accepted

problem3_samples_accept_reject, problem3_acceptance_rate
```

# Part 3

Double click this cell and write you answer for part 3 below, explain what proposal distibution you chose and why you can choose it.

For Part 3, the Uniform distribution was chosen as the proposal distribution for the Accept-Reject sampling method. The reasons for this choice are as follows:

\begin{enumerate}
    \item \textbf{Simplicity and Ease of Sampling}: The Uniform distribution is straightforward to sample from and does not require complex calculations. This simplicity is beneficial when implementing the Accept-Reject method.

    \item \textbf{Coverage of the Target Distribution's Support}: The target distribution \( F[x] \) is non-zero only in the interval \( (0, \ln(2)) \). By choosing a Uniform distribution over the same interval \( [0, \ln(2)] \), it was ensured that the proposal distribution covers the entire range where the target distribution is defined and non-zero.

    \item \textbf{Efficiency}: While the acceptance rate might not be as high as it could be with a more complex proposal distribution, the Uniform distribution still provides a reasonable balance between simplicity and efficiency. In this case, the acceptance rate was approximately 71.17\%, which is quite efficient.

    \item \textbf{Theoretical Justification}: The Accept-Reject method requires that the proposal distribution should be such that it envelopes the target distribution (i.e., the target density should be less than or equal to the scaled proposal density at all points). The Uniform distribution satisfies this condition for the given target distribution \( F[x] \).

    \item \textbf{Ease of Analysis}: The Uniform distribution makes it easier to analyze and understand the sampling process, which is particularly useful for educational purposes or when explaining the methodology to others.
\end{enumerate}

In summary, the Uniform distribution was chosen as the proposal distribution due to its simplicity, coverage of the target distribution's support, reasonable efficiency, theoretical suitability, and ease of analysis.
