21. (Computer Experiment.) Suppose a coin has probability p of falling heads up. If we flip the coin many times, we would expect the proportion of heads to be near p. We will make this formal later. Take p = .3 and n = 1, 000 and simulate n coin flips. Plot the proportion of heads as a function of n. Repeat for p = .03.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the probabilities and the number of coin flips
probabilities = [0.3, 0.03]
n_flips = 1000

# Initialize arrays to store the proportion of heads for each probability
proportions = []

# Simulate coin flips and calculate the proportion of heads
for p in probabilities:
    heads_count = 0
    proportion_heads = []
    
    for _ in range(n_flips):
        if np.random.rand() < p:
            heads_count += 1
        proportion_heads.append(heads_count / (len(proportion_heads) + 1))
    
    proportions.append(proportion_heads)

# Create plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, n_flips + 1), proportions[0], label='p = 0.3')
plt.xlabel('Number of Flips')
plt.ylabel('Proportion of Heads')
plt.title('Proportion of Heads vs. Number of Flips (p = 0.3)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, n_flips + 1), proportions[1], label='p = 0.03', color='orange')
plt.xlabel('Number of Flips')
plt.ylabel('Proportion of Heads')
plt.title('Proportion of Heads vs. Number of Flips (p = 0.03)')
plt.legend()

plt.tight_layout()
plt.show()
```

22. (Computer Experiment.) Suppose we flip a coin n times and let p denote the probability of heads. Let X be the number of heads. We call X a binomial random variable, which is discussed in the next chapter. Intuition suggests that X will be close to np. To see if this is true, we can repeat this experiment many times and average the X values. Carry out a simulation and compare the average of the Xs to np. Try this for p = .3 and n = 10, n = 100, and n = 1, 000.

```python
# Define the probabilities and the number of coin flips (n)
probabilities = 0.3
n_values = [10, 100, 1000]

# Number of simulations
num_simulations = 10000

# Initialize an array to store the average X values for each n
average_X_values = []

for n in n_values:
    X_values = np.random.binomial(n, probabilities, num_simulations)
    average_X = np.mean(X_values)
    average_X_values.append(average_X)

# Calculate np for comparison
np_values = [n * probabilities for n in n_values]

# Print and compare the results
for i, n in enumerate(n_values):
    print(f"For n = {n}, average X: {average_X_values[i]}, np: {np_values[i]}")
```

23. (Computer Experiment.) Here we will get some experience simulating conditional probabilities. Consider tossing a fair die. Let A = {2, 4, 6} and B = {1,2,3,4}. Then, P(A) = 1/2, P(B) = 2/3 and P(AB) = 1/3. Since P(AB) = P(A)P(B), the events A and B are independent. Simu- late draws from the sample space and verify that P(AB) = P(A)P(B) where P(A) is the proportion of times A occurred in the simulation and similarly for P(AB) and P(B). Now find two events A and B that are not independent. Compute P(A), P(B) and P(AB). Compare the calculated values to their theoretical values. Report your results and interpret.

```python
import numpy as np

# Number of simulations
num_simulations = 10000

# Define the sample space (a fair six-sided die)
sample_space = [1, 2, 3, 4, 5, 6]

# Define events A and B
event_A = [2, 4, 6]
event_B = [1, 2, 3, 4]

# Initialize counters for occurrences of A, B, and AB
count_A = 0
count_B = 0
count_AB = 0

# Perform simulations
for _ in range(num_simulations):
    outcome = np.random.choice(sample_space)  # Simulate rolling a die
    
    if outcome in event_A:
        count_A += 1
    if outcome in event_B:
        count_B += 1
    if outcome in event_A and outcome in event_B:
        count_AB += 1

# Calculate the proportions
p_simulated_A = count_A / num_simulations
p_simulated_B = count_B / num_simulations
p_simulated_AB = count_AB / num_simulations

# Calculate the theoretical probabilities
p_theoretical_A = len(event_A) / len(sample_space)
p_theoretical_B = len(event_B) / len(sample_space)
p_theoretical_AB = p_theoretical_A * p_theoretical_B

# Verify independence for events A and B
print("Independence Check:")
print("P(A) * P(B) =", p_theoretical_A * p_theoretical_B)
print("P(AB) =", p_theoretical_AB)
print("P_simulated(AB) =", p_simulated_AB)

# Now, let's find two events that are not independent.
# Let A = {1, 2, 3} and B = {2, 3, 4}.
event_A = [1, 2, 3]
event_B = [2, 3, 4]

# Reset counters
count_A = 0
count_B = 0
count_AB = 0

# Perform simulations
for _ in range(num_simulations):
    outcome = np.random.choice(sample_space)
    
    if outcome in event_A:
        count_A += 1
    if outcome in event_B:
        count_B += 1
    if outcome in event_A and outcome in event_B:
        count_AB += 1

# Calculate the proportions
p_simulated_A = count_A / num_simulations
p_simulated_B = count_B / num_simulations
p_simulated_AB = count_AB / num_simulations

# Calculate the theoretical probabilities
p_theoretical_A = len(event_A) / len(sample_space)
p_theoretical_B = len(event_B) / len(sample_space)
p_theoretical_AB = p_theoretical_A * p_theoretical_B

# Check for independence
print("\nNon-Independence Check:")
print("P(A) * P(B) =", p_theoretical_A * p_theoretical_B)
print("P(AB) =", p_theoretical_AB)
print("P_simulated(AB) =", p_simulated_AB)
```

18. Let X  N (3, 16). Solve the following using the Normal table and using a computer package.
(a) Find P(X < 7). (b) Find P(X > 2).
(c) Find x such that P(X > x) = .05. 
(d) FindP(0X<4).
(e) Find x such that P(|X| > |x|) = .05.

```python
#a
import scipy.stats as stats

mean = 3
std_dev = 4  # Square root of variance

# Using the CDF to find P(X < 7)
probability = stats.norm.cdf(7, loc=mean, scale=std_dev)
print("P(X < 7) =", probability)
```

```python
#b
# Using the CDF to find P(X > -2)
probability = 1 - stats.norm.cdf(-2, loc=mean, scale=std_dev)
print("P(X > -2) =", probability)
```

```python
#c
# Using the percent-point function (PPF) to find the value of x
x = stats.norm.ppf(0.05, loc=mean, scale=std_dev)
print("x such that P(X > x) = 0.05 is x =", x)
```

```python
#d
# Using the CDF to find P(0  X < 4)
probability = stats.norm.cdf(4, loc=mean, scale=std_dev) - stats.norm.cdf(0, loc=mean, scale=std_dev)
print("P(0  X < 4) =", probability)
```

```python
#e
# Using the percent-point function (PPF) to find the value of x
x = abs(stats.norm.ppf(0.025, loc=mean, scale=std_dev))  # 0.025 on each tail
print("x such that P(|X| > |x|) = 0.05 is x =", x)
```

9. (Computer Experiment.) Let X1,X2,...,Xn be N(0,1) random variables and let Xn = n1 ni=1 Xi. Plot Xn versus n for n = 1,...,10,000. Repeat for X1, X2, . . . , Xn  Cauchy. Explain why there is such a dif- ference.

```python
# Number of terms to sum (n)
n_values = np.arange(1, 10001)

# Initialize arrays to store the results for standard normal and Cauchy distributions
Xn_normal = []
Xn_cauchy = []

# Simulate Xn for each n
for n in n_values:
    Xi_normal = np.random.standard_normal(n)  # Generating n standard normal random variables
    Xn_normal.append(np.sum(Xi_normal) / np.sqrt(n))  # Xn = (X1 + X2 + ... + Xn) / sqrt(n)
    
    Xi_cauchy = np.random.standard_cauchy(n)  # Generating n Cauchy random variables
    Xn_cauchy.append(np.sum(Xi_cauchy) / n)  # Xn = (X1 + X2 + ... + Xn) / n

# Create plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(n_values, Xn_normal)
plt.xlabel('n')
plt.ylabel('Xn')
plt.title('Xn vs. n (Standard Normal Distribution)')

plt.subplot(1, 2, 2)
plt.plot(n_values, Xn_cauchy)
plt.xlabel('n')
plt.ylabel('Xn')
plt.title('Xn vs. n (Cauchy Distribution)')

plt.tight_layout()
plt.show()
```

11. (Computer Experiment: Simulating the Stock Market.) Let Y1,Y2,... be independent random variables such that P(Yi = 1) = P(Yi = 1) = 1/2. Let Xn = ni=1 Yi. Think of Yi = 1 as the stock price increased by one dollar, Yi = 1 as the stock price decreased by one dollar, and Xn as the value of the stock on day n.
(a) Find E(Xn) and V(Xn).
(b) Simulate Xn and plot Xn versus n for n = 1,2,...,10,000. Repeat the whole simulation several times. Notice two things. First, its easy to see patterns in the sequence even though it is random. Second, you will find that the four runs look very different even though they were generated the same way. How do the calculations in (a) explain the second observation?

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of simulations
num_simulations = 4

# Number of days
n_values = np.arange(1, 10001)

# Initialize an array to store the results for each simulation
Xn_simulations = []

# Simulate Xn for each run
for _ in range(num_simulations):
    Y_values = np.random.choice([-1, 1], size=len(n_values), p=[0.5, 0.5])
    Xn_values = np.cumsum(Y_values)
    Xn_simulations.append(Xn_values)

# Create plots for each simulation
plt.figure(figsize=(12, 6))
for i, Xn_values in enumerate(Xn_simulations):
    plt.plot(n_values, Xn_values, label=f"Run {i + 1}")

plt.xlabel('n (Days)')
plt.ylabel('Xn (Stock Value)')
plt.title('Simulated Stock Market (Xn) for Different Runs')
plt.legend()
plt.grid(True)

plt.show()
```
